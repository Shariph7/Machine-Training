```python
# Importing the necessary Libraries
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
# Loading the datasets
data = pd.read_csv('../Supervised DataSets/MultipleRegression.csv')
data.head()
```

```python
# EDA(Exploratory Data Analysis)
data.info()
```

```python
data.describe()
```

```python
# Visualization
import matplotlib.pyplot as plt
v = data['Volume']
c = data['CO2']
plt.plot(v, c)
```

```python
# Making Model
X = data[['Weight', 'Volume']]
y = data['CO2']
```

```python
# Preprocessing the Data
scale = StandardScaler()
```

```python
X_scaled = scale.fit_transform(X)
```

```python
X_scaled
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_scaled, y, test_size = 0.2, random_state = 42)
```

```python
# Using the Linear Regression Model
regr = LinearRegression()
regr.fit(xtrain, ytrain)
```

```python
# Checking Accuracy of the Model
y_pred = regr.predict(xtest)
accuracy = r2_score(ytest, y_pred)
```

```python
# Accuracy of the model
print(f"The Accuracy of the Model is {((accuracy)*100):.2f} %")
```

```python
scaled = scale.transform([[2300, 1300]])
```

```python
# Predicting the Amount of CO2 Emission
predictedCO2 = regr.predict([scaled[0]])
```

```python
print(f"The Amount of CO2 Emission is {predictedCO2}")
```
